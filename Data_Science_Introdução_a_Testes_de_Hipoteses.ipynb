```python
from statsmodels.stats.weightstats import ztest

ztest(notas1['rating'], value = nota_media_dos_filmes_com_pelo_menos_10_votos.mean())
```

8.978190401885648e-18 Ã© um valor minusculo, entÃ£o a media do Toy Story nÃ£o pode ser igual a media dos filmes no geral.
Rejeitamos a hipotese de Toy Story, no mundo real, ter a mesma media que a media de todos os filmes.

```python
np.random.seed(75241)
temp = notas1.sample(frac=1)['rating']

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value = nota_media_dos_filmes_com_pelo_menos_10_votos.mean())
  return(i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:,0], valores[:,1])
#eixo x = valores[:,0] Quantidade de Amostras
#eixo y = valores[:,1] Media das notas

plt.plot(valores[:,0], valores[:,2])
#eixo x = valores[:,0] Quantidade de Amostras
#eixo y = valores[:,2] P value
plt.hlines(y = 0.05, xmin= 2, xmax = len(temp), colors='g')
```

A visualizaÃ§Ã£o do P value em grafico nÃ£o Ã© comum, mas Ã© bom para entender que dependendo da quantidade de valores na amostra poderiamos ter entendido que a nossa hipotese(Toy Story ter a mesma media de notas que os filmes no geral) seria aceita, atÃ© os primeiros 45 registros. Como temos muitos dados e olhamos para eles, vemos que essa hipotese nÃ£o Ã© real, descartamos ela.

Ou seja essa normalizaÃ§Ã£o se espera que os dados venham de uma base de dados que represente uma normal, como o nosso tem uma assimetria a esquerda, ele descarta a hipotese nula(0.00011053430732728716

```python
from scipy.stats import ranksums

_, p = ranksums(notas1['rating'], notas593['rating'])
p
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ranksums.html

"The Wilcoxon rank-sum test tests the **null hypothesis that two sets of measurements are drawn from the same distribution**. The **alternative hypothesis is that values in one sample are more likely to be larger than the values in the other sample**."

Vamos reparar que nÃ£o fala sobre media, aqui esse metodo nos diz que geralmente as notas aleatores de um filme (Silence of the Lambs) serÃ£o maiores que a de outro (Toy Story), que foram os nossos casos.

EntÃ£o nesta hipotese de notas maiores nÃ³s aceitamos a hipose alternativa. Por que o valor 0.0003267718756440693 Ã© menor do que 0.5 se fosse maior aceitariamos a nula.

PS: Como nÃ£o sabemos qual Ã© a distribuiÃ§Ã£o correta para os nossos dados e sabemos que ele nÃ£o Ã© uma distribuiÃ§Ã£o normal, nÃ³s rejeitamos a hipotese nula.
