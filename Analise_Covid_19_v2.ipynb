A ideia era corrigir aquela conta e pegar o valor real da populaÃ§Ã£o porÃ©m estamos com uma dificuldade nosso dataframe inicial nÃ£o tinha essa informaÃ§Ã£o, fomos em uma nova base de dados que contem essa informaÃ§Ã£o: https://population.un.org/wpp/Download/Standard/CSV/

PorÃ©m o nome dos paises nÃ£o sÃ£o iguais, vou assumir que United States of America Ã© igual ao US e seguir a vida.

```python
def rename_location(location):
  if location == 'United States of America':
    return 'US'
  if location == 'South, Korea':
    return 'Korea, South'
  if location == 'Iran (Islamic Republic of)':
    return 'Iran'
  if location == 'Russian Federation':
    return 'Russia'
  if location == 'China, Taiwan Province of China':
    return 'Taiwan*'
  return location

un_population_total['location_for_who'] = un_population_total.index.map(rename_location)
un_population_total.head()
```

```python
un_population_total.query('location_for_who == "Taiwan*"').head()
```

```python
un_population_total = un_population_total.set_index('location_for_who')
un_population_total.head()
```

```python
combined.join(un_population_total).sort_values('confirmed', ascending=False).head(20)
```

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head(10)
```
